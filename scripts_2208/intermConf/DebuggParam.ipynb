```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
as_nblw = 1567.8
bs_nblw = -18.975
cs_nblw = -17.668
#R2
am_nblw = 1697.1
bm_nblw = -15.311
cm_nblw = -64.153

#Values for parametrization for eta > 0.8
#R1
as_nabv = 1503.2
bs_nabv = 71.716
cs_nabv = -41.008
#R2
am_nabv = 1739.1
bm_nabv = -75.648
cm_nabv = -18.501
```

```python
eta = np.linspace(0, 0.8, 100)
ys1 = as_nblw + bs_nblw*eta + cs_nblw*(eta**2)
ys2 = am_nblw + bm_nblw*eta + cm_nblw*(eta**2)

eta2 = np.linspace(0.8, 1.4, 100)
ys3= as_nabv + bs_nabv*eta2 + cs_nabv*(eta2**2)
ys4= am_nabv + bm_nabv*eta2 + cm_nabv*(eta2**2)

plt.plot(eta, ys1, label='R1 for eta < 0.8', color='red')
plt.plot(eta, ys2, label='R2 for eta < 0.8', color='blue')

plt.plot(eta2, ys3, label='R1 for eta > 0.8', color='green')
plt.plot(eta2, ys4, label='R2 for eta > 0.8', color='orange')

plt.legend()
```

```python
def R2plus(eta22):
    ys4= am_nabv + bm_nabv*eta22 + cm_nabv*(eta22**2)
    return ys4

def R1plus(eta22):
    ys3= as_nabv + bs_nabv*eta22 + cs_nabv*(eta22**2)
    return ys3

def R1minus(eta22):
    ys1 = as_nblw + bs_nblw*eta22 + cs_nblw*(eta22**2)
    return ys1

def R2minus(eta22):
    ys2 = ys2 = am_nblw + bm_nblw*eta22 + cm_nblw*(eta22**2)
    return ys2
```
