```python
def read_dataLess300(file_path):
    x_values = []
    y_values = []
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.split()
            if len(parts) == 4:
                x = np.abs(float(parts[1]))
                constant = float(parts[2])
                y = float(parts[3])
                if x < 300:
                    x_values.append(x + constant)
                    y_values.append(y)

    return x_values, y_values
```

```python
def read_dataMore300(file_path):
    x_values = []
    y_values = []
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.split()
            if len(parts) == 4:
                x = np.abs(float(parts[1]))
                constant = float(parts[2])
                y = float(parts[3])
                if x > 300:
                    x_values.append(x + constant)
                    y_values.append(y)

    return x_values, y_values
```

```python
def read_dataAtlas_complete(file_path):
    x_values = []
    y_values = []
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.split()
            if len(parts) == 4:
                x = np.abs(float(parts[0]))
                y = float(parts[3])
                x_values.append(x)
                y_values.append(y)

    return x_values, y_values
```

```python
def Atlas_vs_Simple(file_path):
    x_values = []
    y_values = []
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.split()
            if len(parts) == 4:
                x = float(parts[0])
                y = float(parts[1])
                x_values.append(x)
                y_values.append(y)

    return x_values, y_values
```

```python
def compare(file_path):
    x_values = []
    y_values = []
    zsimp300_zatlas300 = 0
    zsimp300_zatlas = 0
    zsimp_zatlas300 = 0
    with open(file_path, 'r') as file:
        for line in file:
            parts = line.split()
            if len(parts) == 4:
                zatlas = float(parts[0]) 
                zsimp = float(parts[1])
                
                if (zsimp>300. and zatlas>300.):                    
                    zsimp300_zatlas300+=1
                if(zsimp>300. and zatlas<300.):
                    zsimp300_zatlas+=1
                if (zsimp<300. and zatlas>300.):
                    zsimp_zatlas300+=1

    return zsimp300_zatlas300, zsimp300_zatlas, zsimp_zatlas300
```
